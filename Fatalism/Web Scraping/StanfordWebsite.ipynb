```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = "https://plato.stanford.edu/entries/fatalism/"
response = requests.get(url)
print(response.text)
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)

info = html_soup.find_all('p')
print(type(info))
print(len(info))
```

```python
print(info)
```

```python
article_text = ''

info_div = html_soup.find("div").findAll('p')
for element in info_div:
    article_text += '\n' +''.join(element.findAll(text=True))
```

```python
article_text = article_text.replace("\n", " ")
#article_text = article_text.replace('\xe2\x80\x9', " ")
#article_text = article_text.replace("'", " ")
article_text = article_text.replace("(", " ")
article_text = article_text.replace(")", " ")
article_text = article_text.replace("[", " ")
article_text = article_text.replace("]", " ")

article_text = article_text.encode('ascii', 'replace')
```

```python
print(article_text)
```

```python
import sys

print(sys.stdout.encoding)

article_text = str(article_text)
file_name = 'data.txt'

file = open(file_name, 'w')
file.write(article_text)
```
